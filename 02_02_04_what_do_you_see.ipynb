```python
# import NBA player of the week data to pandas dataframe
```

```python
df_nba = pd.read_csv('NBA_player_of_the_week.csv')
```

```python
print(df_nba)
```

```python
player_age = df_nba["Age"]
```

```python
print(player_age)
```

```python
# view player ages in a boxplot
```

```python
plt.boxplot(player_age) 
plt.title('Ages of NBA Player of the Weeks')
plt.ylabel('Age')
plt.show()
```

From this graph we can see the average age for the NBA player
of the week varies between 19 and 40.

The lowest value to the 25th percentile is about 19-24.

The 25th to 75th percentile is about 24-29 years of age.

The median age is about 26 years old.  

It looks like ages 37-40 are outliers. This was unexpected because 40 does not appear to be any more of an outlier than 19.  

```python
# view same player ages in a histogram
```

```python
plt.hist(player_age)
plt.title('Ages of NBA Player of the Weeks')
plt.xlabel('Age')
plt.title('Ages of NBA Player of the Weeks')
plt.show()
```

Looking at the same data in a histogram is consistent with the boxplot view.

However, this helps me to see why 40 was an outlier but 19 was not.

I see everything is much closer to 19 than it is to 40.

```python
# Subplot example with the NBA data set
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(df_nba['Age'], color='purple')
plt.ylabel('Age')
plt.title('Age')

plt.subplot(2, 1, 2)
plt.plot(df_nba['Seasons in league'], color='green')
plt.ylabel('Years')
plt.title('Years in League')
plt.show()
```

As expected, these two graphs are somewhat similar. Players who are younger will generally have fewer years in the league and players who are older will generally have more years in the league.
